```python
import geopandas as gpd
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gdf = gpd.read_file("data/districts/districts.shp")
climate_df = pd.read_csv("data/monthly_climate.csv")
food_df = pd.read_csv("data/wfp_food_prices_npl.csv", low_memory=False)
food_df = food_df.drop(index=0).reset_index(drop=True)

# Rename columns
food_df = food_df.rename(columns={"admin2": "DISTRICT"})
climate_df = climate_df.rename(columns={"District": "DISTRICT"})

# Lowercase for uniformity
food_df['DISTRICT'] = food_df['DISTRICT'].str.strip().str.lower()
climate_df['DISTRICT'] = climate_df['DISTRICT'].str.strip().str.lower()
gdf['DISTRICT'] = gdf['DISTRICT'].str.strip().str.lower()

# fix district name inconsistency
shape_name_corrections = {
    "rukum" : ["rukum_e", "rukum_w"],
    "nawalparasi" : ["nawalparasi_w", "nawalparasi_e"]
}

# Flatten and invert for .replace()
shape_flattened = {sub: main for main, subs in shape_name_corrections.items() for sub in subs}
gdf['DISTRICT'] = gdf['DISTRICT'].replace(shape_flattened)

# Dissolve shapefile
gdf = gdf.dissolve(by='DISTRICT', as_index=False)

# List of columns to drop
columns_to_drop = [
    "WetBulbTemp_2m", #correlated with RH + Temp
    "EarthSkinTemp", #correlated with AirTemp
    "MaxWindSpeed_10m",
    "MinWindSpeed_10m",
    "WindSpeedRange_10m",
    "MaxWindSpeed_50m",
    "MinWindSpeed_50m",
    "WindSpeedRange_50m"
]
climate_df = climate_df.drop(columns=[col for col in columns_to_drop if col in climate_df.columns])

climate_name_corrections = {
    "dhanusa" : "dhanusha",
    "kabhre" : "kabhrepalanchok",
    "routahat" : "rautahat",
    "tanahun" : "tanahu",
    "dhanusa" : "dhanusha",
    "makwanpur" : "makawanpur",
    "bajang" : "bajhang",
    "dolkha" : "dolakha",
    "panchther" : "panchthar"
}

climate_df['DISTRICT'] = climate_df['DISTRICT'].replace(climate_name_corrections)

food_name_corrections = {
    "rukum_w" : "rukum",
    "kavre" : "kabhrepalanchok",
}

food_df['DISTRICT'] = food_df['DISTRICT'].replace(food_name_corrections)

# Unique districts
climate_districts = set(climate_df['DISTRICT'])
food_districts = set(food_df['DISTRICT'])
shape_districts = set(gdf['DISTRICT'])

# Compare
print(climate_df.info())
print(food_df.info())
```

```python
# Standardize and fix date format
climate_df['date'] = pd.to_datetime(climate_df['date'])
food_df['date'] = pd.to_datetime(food_df['date'])

# Replace day with 15 to match monthly format
climate_df['date'] = climate_df['date'].apply(lambda d: d.replace(day=15))
food_df['date'] = food_df['date'].apply(lambda d: d.replace(day=15))

# Filter date range
start_date = "2001-04-15"
end_date = "2019-12-15"
climate_df = climate_df[(climate_df["date"] >= start_date) & (climate_df["date"] <= end_date)]
food_df = food_df[(food_df["date"] >= start_date) & (food_df["date"] <= end_date)]

# Filter for selected commodities
selected_commodities = ["Rice (coarse)", "Wheat flour"]
food_df = food_df[food_df["commodity"].isin(selected_commodities)]

# Convert price to numeric, coercing errors
food_df['price'] = pd.to_numeric(food_df['price'], errors='coerce')

# Pivot food prices
food_pivot = food_df.pivot_table(
    index=["DISTRICT", "date"],
    columns="commodity",
    values="price",
    aggfunc="mean"
).reset_index()
print(f"Pivot table created with {len(food_pivot)} rows")

# Merge
merged_df = pd.merge(
    climate_df,
    food_pivot,
    how="inner",
    on=["DISTRICT", "date"]
)

merged_df = merged_df.rename(columns={"Rice (coarse)": "Rice(coarse)_price", "Wheat flour": "Wheat_flour_price"})

merged_df['Rice(coarse)_price'] = merged_df.groupby('DISTRICT')['Rice(coarse)_price'].transform(lambda x: x.interpolate(method='linear').ffill().bfill())
merged_df['Wheat_flour_price'] = merged_df.groupby('DISTRICT')['Wheat_flour_price'].transform(lambda x: x.interpolate(method='linear').ffill().bfill())

# Save to CSV
merged_df.to_csv("merged_climate_food.csv", index=False)
print(merged_df[['Rice(coarse)_price', 'Wheat_flour_price']].isna().sum())
print(merged_df.head())
```

```python
# price trends over time:
plt.figure(figsize=(14, 6))
sns.lineplot(data=merged_df, x='date', y='Rice(coarse)_price', hue='DISTRICT')
plt.title('Rice (Coarse) Price Trends by District')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(14, 6))
sns.lineplot(data=merged_df, x='date', y='Wheat_flour_price', hue='DISTRICT')
plt.title('Wheat Flour Price Trends by District')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of Prices
sns.histplot(merged_df['Rice(coarse)_price'], kde=True)
plt.title("Distribution of Rice (Coarse) Price")
plt.show()

sns.histplot(merged_df['Wheat_flour_price'], kde=True)
plt.title("Distribution of Wheat Flour Price")
plt.show()
```
